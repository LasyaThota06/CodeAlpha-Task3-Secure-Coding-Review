```python
username = input("Username: ")
password = input("Password: ")

if username == "admin" and password == "admin123":
    print("Login Successful")
else:
    print("Login Failed")
```

```python
print("SECURE CODING REVIEW")

print("\nVulnerability 1: Hardcoded Credentials")
print("Risk: High")
print("Credentials are stored directly in the source code.")

print("\nVulnerability 2: Plaintext Password")
print("Risk: High")
print("Password is compared in plain text.")

print("\nVulnerability 3: Unlimited Login Attempts")
print("Risk: Medium")
print("No protection against brute-force attacks.")

print("\nRecommendations:")
print("- Use password hashing")
print("- Store credentials securely")
print("- Limit failed login attempts")
```
